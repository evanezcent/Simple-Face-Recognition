```python
import numpy as np
import cv2
# print(cv2.__file__)
import pickle
```

```python
face_cascade = cv2.CascadeClassifier('C:cascades/data/haarcascade_frontalface_alt2.xml')
eye_cascade = cv2.CascadeClassifier('C:cascades/data/haarcascade_eye.xml')
smile_cascade = cv2.CascadeClassifier('C:cascades/data/haarcascade_smile.xml')
recognizer = cv2.face.LBPHFaceRecognizer_create()
recognizer.read("trainner.yml")

og_labels = {}
labels = {}
with open("labels.pickle", "rb") as f:
    og_labels = pickle.load(f)
    labels = {v:k for k,v in og_labels.items()}
    print(labels)
#     labels = {v:k for k,v in og_labels.item()}
```

```python
# membuat camera
cam = cv2.VideoCapture(0)

while(True):
#     camera membaca tiap frame / fotonya
    ret, frame = cam.read()
    
#     membuat warna grayscale sesuai kerja CV 
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor = 1.5, minNeighbors = 5)
    
    for (x, y, w, h) in faces :
#         print(x, y, w, h)
        roi_gray = gray[y:y+h, x:x+w] #(ycord_start, ycord_ende)
        roi_color = frame[y:y+h, x:x+w]
        
#         recognize dengan prediksi model deep learning seperti keras, tensorflow, pytorch, scikit-learn
        id_, conf = recognizer.predict(roi_gray)
        if 45 <= conf:
            print(id_)
            print(labels[id_])
            font = cv2.FONT_HERSHEY_SIMPLEX
            name = labels[id_]
            color = (255, 199, 99)
            stroke = 2 
            cv2.putText(frame, name, (x,y), font, 1, color, stroke, cv2.LINE_AA)
        
        img_item = "7.png"
        cv2.imwrite(img_item, roi_gray)
        
        
        color = (255, 0, 0)
        stroke = 2
        end_cord_x = x + w
        end_cord_y = y + h
        cv2.rectangle(frame, (x, y), (end_cord_x, end_cord_y), color, stroke)   
        
#         detect Eye
        eyes = eye_cascade.detectMultiScale(roi_gray)
        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 100, 30), 2)   
        
#         Detect smile
        smiles = smile_cascade.detectMultiScale(roi_gray)
        for (sx, sy, sw, sh) in eyes:
            cv2.rectangle(roi_color, (sx, sy), (sx+sw, sy+sh), (90, 100, 30), 2)   
        
#     Menampilkan hasil frame
    cv2.imshow('frame', frame)
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break

# Mematikan kamera
cam.realease()
cv2.destroyAllWindows()
```
